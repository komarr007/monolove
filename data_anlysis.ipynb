```python
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    phi = chi2 / n
    r, k = contingency_table.shape
    v = np.sqrt(phi / (min(r, k) - 1))
    return v
```

```python
corr_matrix = pd.DataFrame(index=df_training[['pf_o_sin', 'pf_o_int', 'pf_o_fun', 'pf_o_amb', 'pf_o_sha']].columns, columns=df_training[['pf_o_sin', 'pf_o_int', 'pf_o_fun', 'pf_o_amb', 'pf_o_sha']].columns)
for i in df_training[['pf_o_sin', 'pf_o_int', 'pf_o_fun', 'pf_o_amb', 'pf_o_sha']].columns:
    for j in df_training[['pf_o_sin', 'pf_o_int', 'pf_o_fun', 'pf_o_amb', 'pf_o_sha']].columns:
        corr_matrix.loc[i, j] = cramers_v(df[i], df[j])

# output the correlation matrix
corr_matrix
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(df_training_scaled, test_size=0.20, shuffle=True, random_state=42)
X_validation, X_test = train_test_split(X_test, test_size=0.50, shuffle=True, random_state=42)
```

## classify match

```python
x_train = X_train.drop(['match','iid','pid'],axis=1)
y_train = X_train['match']

x_val = X_validation.drop(['match','iid','pid'],axis=1)
y_val = X_validation['match']
```

### modelling

```python
import tensorflow as tf
```

```python
input_features = []

for key in x_train.columns:
    input_features.append(
        tf.keras.Input(shape=(1,), name=key)
    )
```

```python
x = tf.keras.layers.concatenate(input_features)
x_1 = tf.keras.layers.Dense(16, activation="relu")(x)

output = tf.keras.layers.Dense(1, activation="sigmoid")(x_1)

inputs = input_features

keras_model = tf.keras.models.Model(inputs, output)
keras_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=[
        tf.keras.metrics.BinaryAccuracy(),
        tf.keras.metrics.Precision(),
        tf.keras.metrics.Recall(),
        tf.keras.metrics.AUC(),
    ],
)

keras_model.summary()
```

```python
history = keras_model.fit(
    [x_train[col] for col in x_train.columns],
    y_train,
    epochs=50,
    validation_data=([x_val[col] for col in x_val.columns],y_val),
)
```

```python
for key in history.history:
    print(key)
```

```python
print('score loss train: ', history.history['loss'][-1])
print('score accuracy: ', history.history['binary_accuracy'][-1])
print('score val_loss: ', history.history['val_loss'][-1])
print('score val_accuracy: ', history.history['val_binary_accuracy'][-1])
plt.plot(history.history['loss'][10:])
plt.plot(history.history['binary_accuracy'][10:])
plt.plot(history.history['val_loss'][10:])
plt.plot(history.history['val_binary_accuracy'][10:])
plt.title('model_metrics')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'binary_accuracy', 'val_loss', 'val_binary_accuracy'], loc='upper left')
plt.show()
```

### system recommendation

```python
def output_recommendation(user_id):

    suggested_id = []

    user_data = X_test.loc[(X_test['iid'] == user_id)]

    for ind in user_data.index:
        data = user_data[user_data.index == ind].drop(['pid','iid','match'],axis=1)
        data_orisinil = user_data[user_data.index == ind]
        result = model.predict(data)
        pid_prob = (data_orisinil.pid,result[0])
        suggested_id.append(pid_prob)

    return suggested_id

def print_output(user_id):

    user_data_gender = int(X_test.loc[(X_test['iid'] == user_id)].sample(1).gender)

    list_output = output_recommendation(user_id)

    sorted_list = sorted(list_output, key=lambda x: x[1], reverse=True)

    if user_data_gender == 1:
        print("\nTOP 3 WOMAN PROBABLY MATCH".center(20,'*'))
        for key, vals in sorted_list[:3]:
            print("PID : {} and probability : {}".format(int(key), vals))

    elif user_data_gender == 0:
        print("\nTOP 3 MAN PROBABLY MATCH".center(20,'*'))
        for key, vals in sorted_list[:3]:
            print("PID : {} and probability : {}".format(int(key), vals))

    else:
        print("gender gajelas")
```
